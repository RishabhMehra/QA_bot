```python
import logging

from simpletransformers.question_answering import QuestionAnsweringModel, QuestionAnsweringArgs
```

```python
import json
with open(r"GRADSIL_test.json", "r") as read_file:
    test = json.load(read_file)
```

```python
with open(r"GRADSIL_train.json", "r") as read_file:
    train = json.load(read_file)
```

```python
train
```

```python
model_type="roberta"
model_name= "bert-base-cased"
if model_type == "roberta":
    model_name = "deepset/roberta-base-squad2-covid"

# elif model_type == "roberta":
#     model_name = "roberta-base"

elif model_type == "distilbert":
    model_name = "distilbert-base-cased"

elif model_type == "distilroberta":
    model_type = "roberta"
    model_name = "distilroberta-base"

elif model_type == "electra-base":
    model_type = "electra"
    model_name = "google/electra-base-discriminator"

elif model_type == "electra-small":
    model_type = "electra"
    model_name = "google/electra-small-discriminator"

elif model_type == "xlnet":
    model_name = "xlnet-base-cased"
```

```python
# Configure the model 
model_args = QuestionAnsweringArgs()
model_args.train_batch_size = 16
model_args.evaluate_during_training = True
model_args.n_best_size=8
model_args.num_train_epochs=10
```

```python
### Advanced Methodology
train_args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "use_cached_eval_features": True,
    "output_dir": f"outputs/{model_type}",
    "best_model_dir": f"outputs/{model_type}/best_model",
    "evaluate_during_training": True,
    "max_seq_length": 512,
    "num_train_epochs": 10,
    "evaluate_during_training_steps": 1000,
    "wandb_project": "Question Answer Application",
    "wandb_kwargs": {"name": model_name},
    "save_model_every_epoch": False,
    "save_eval_checkpoints": False,
    "n_best_size":8,
    # "use_early_stopping": True,
    # "early_stopping_metric": "mcc",
    # "n_gpu": 2,
    # "manual_seed": 4,
    # "use_multiprocessing": False,
    "train_batch_size": 256,
    "eval_batch_size": 64,
    # "config": {
    #     "output_hidden_states": True
    # }
}
```

```python
model = QuestionAnsweringModel(
    model_type,model_name, args=train_args
)
```

```python
# Train the model
model.train_model(train, eval_data=test)
```

```python
# Evaluate the model
result, texts = model.eval_model(test)
```

```python
# Make predictions with the model
to_predict = [
    {
        "context": "During its February 2015 meeting, the Advisory Committee on Immunization Practices (ACIP) recommended 9-valent human papillomavirus (HPV) vaccine (9vHPV) (Gardasil 9, Merck and Co., Inc.) as one of three HPV vaccines that can be used for routine vaccination (Table 1).",
        "qas": [
            {
                "question": "When did the GARDASIL 9 recommendations change?",
                "id": "0",
            }
        ],
    }
]
```

```python
answers, probabilities = model.predict(to_predict)

print(answers)
```
